```python
# coding:utf-8

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
trade = pd.read_csv('../datas/lgb60_month2_yval_pred')
trade.head()
```

```python
trade[trade['is_risk']==1]['y'].plot.hist(200,figsize=(16,4))
plt.show()
```

```python
trade[trade['is_risk']==0]['y'].plot.hist(100,figsize=(16,4))
plt.show()
```

```python
p1 = trade[(trade['is_risk']==1)&(trade['y']<0.2)]
dpa = p1[['id','is_risk']].groupby('id',as_index=False).sum()
dpa.sort_values('is_risk',ascending=False).head(50)
```

```python
p0 = trade[(trade['is_risk']==0)&(trade['y']>0.3)]
dpb = p0[['id','is_risk']].groupby('id',as_index=False).count()
dpb.sort_values('is_risk',ascending=False).head(50)
```
