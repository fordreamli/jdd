```python
# coding:utf-8

import pandas as pd
import numpy as np
import os,sys 
from multiprocessing import Pool
import matplotlib.pyplot as plt
```

```python
xgb100_val = pd.read_csv('../datas/xgb100_yval_pred')
xgb100_val.head()
```

```python
xgb100_val['a'] = xgb100_val['rowkey'].map(lambda x:1 if x in a else 0)
xgb100_val.head(5)
```

```python
xgb100_val = xgb100_val.sort_values(['a'],ascending=False).reset_index(drop=True)
xgb100_val.head(5)
```

```python
xs = xgb100_val[xgb100_val['is_risk']==1].shape[0]
n = 0
def f(x):
    global n
    n += x
    return n
xgb100_val['cnt'] = xgb100_val['is_risk'].map(f)
xgb100_val['recall'] = 1.0*xgb100_val['cnt']/xs
xgb100_val['precision'] = 1.0*xgb100_val['cnt']/(1+xgb100_val.index)
xgb100_val['f'] =(1+0.1*0.1)*xgb100_val['precision']*xgb100_val['recall']/(0.1*0.1*xgb100_val['precision']+xgb100_val['recall'])
xgb100_val.head()
```

```python
xgb100_val['rang'] = xr
xgb100_val.head()
```
